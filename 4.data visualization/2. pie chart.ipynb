```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

```python
co=['w','b','black']
plt.pie(y,colors=co,labels=["DSP","CD","COA"])
plt.show()
```

```python
plt.legend(title="subject")
plt.show()
```

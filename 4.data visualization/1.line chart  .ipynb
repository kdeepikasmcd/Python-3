```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xaxis=np.array([0,20])
yaxis=np.array([0,150])
plt.plot(xaxis,yaxis)        #line is diagram from position(0,0) to (20,150)
print(plt.show())
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints=np.array([1,200])
ypoints=np.array([5,150])
plt.plot(xpoints,ypoints)
print(plt.show())
```

```python
plt.plot([5,100],[5,200],"--")
plt.show()
```

```python
#multiple points
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints)
plt.scatter(xpoints,ypoints)
plt.show()
```

```python
#plotting without xpoints
ypoints=np.array([1,3,8,10,5,0])
plt.plot(ypoints)
plt.show()
```

```python
#plotting without ypoints
xpoints=np.array([2,4,5,63,66,2])
plt.plot(xpoints)
plt.show()
```

```python
#plotting without xpoints
ypoints=np.array([1,3,4,7,2,10])
plt.plot(xpoints,marker="*",mec='r') #instead of * =. or ,
plt.show()
```

```python
ypoints=np.array([2,4,5,8,9])
plt.plot(ypoints,marker="o",mec="b", color="r")
plt.show()
```

```python
ypoints=np.array([2,4,5,8,9])
plt.plot(ypoints,marker="o",mec="b", color="r",linewidth="10")
plt.show()
```

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints,marker="o",mec="b", color="black",linewidth="2")
plt.show()
```

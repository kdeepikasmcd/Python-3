```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
x=np.random.rand(4)
y=np.random.rand(4)
plt.scatter(x,y)
plt.show()
```

```python
plt.title("line plot")
plt.xlabel("x-axixs")
plt.ylabel("y-axis")
x=[1,3,4,6,8]
y=[2,4,6,8,9]
plt.scatter(x,y)
print(plt.show())
```

```python
plt.scatter(x,y,c='r')
```

```python
color=['red','green','blue','pink','orange']
plt.scatter(x,y,c=color)
plt.show()
```

```python
plt.scatter(x,y,s=100)
plt.show()
```

```python
size=1,10,20,50,100
plt.scatter(x,y,s=size)
plt.show()
```

```python
plt.scatter(x,y,marker="*")
plt.show()
```

```python
mark=['1','*','^','.','s']
plt.scatter(x,y,marker=mark)
plt.show()
```

```python
plt.scatter(x,y,edgecolor="pink")
plt.show()
```

```python
plt.scatter(x,y,linewidth="10")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.rand(3)
y=np.random.rand(3)
z=np.random.rand(3)
colors=np.random.rand(3)
plt.scatter(x,y,c=colors,s=z*100)
plt.show()
```

```python
plt.scatter(x,y,marker="*",c='green',alpha=1,s=4,edgecolor='green',linewidth=3)
plt.show()
```

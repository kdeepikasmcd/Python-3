```python
import matplotlib.pyplot as plt
country=["India","china","Japan","Germany","France"]
gdp=[30000,25000,41000,29000,20000]
plt.bar(country,gdp,color="black")
plt.title("country vs gdp per capita")
plt.xlabel("country")
plt.ylabel("gdp per capita")
plt.show()
```
